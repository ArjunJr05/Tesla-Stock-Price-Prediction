```python
layout = go.Layout(
    title='Stock Prices of Tesla',
    xaxis=dict(
        title='Date',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Price',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

tesla['Date'] = pd.to_datetime(tesla['Date'])
tesla_dates = np.array(tesla['Date'])

tesla_data = go.Scatter(x=tesla_dates, y=tesla['Close'], mode='lines', name='Tesla Stock Price')

plot = go.Figure(data=[tesla_data], layout=layout)


```
